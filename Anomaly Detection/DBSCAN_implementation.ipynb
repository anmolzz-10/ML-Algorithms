```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn.cluster import DBSCAN 
```

```python
make_circles(n_samples= 750)
```

```python
X,y = make_circles(n_samples = 750,factor = 0.3 , noise = 0.1)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
dbscan = DBSCAN(eps=0.10)
```

```python
predictions = dbscan.fit_predict(X)
predictions
```

```python
import numpy as np
index = np.where(predictions<0)
index
```

```python
plt.scatter(X[:,0],X[:,1],c=predictions) ##dbscab.labels__ can also be used besides predictions variable 
plt.scatter(X[index,0],X[index,1], c ='r') ##outlier denoted by red circles
```

```python
##compared to real data points

plt.scatter(X[:,0],X[:,1],c=y)
```

## Local Outlier Factor

```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(n_neighbors= 20,contamination= 0.05 )
```

```python
predictions = lof.fit_predict(X)
```

```python
import numpy as np
index = np.where(predictions<0)
index
```

```python
plt.scatter(X[:,0],X[:,1])  
plt.scatter(X[index,0],X[index,1], c ='r') 
```

```python
import matplotlib.pyplot as plt

inliers = X[predictions == 1]
outliers = X[predictions == -1]

plt.scatter(inliers[:, 0], inliers[:, 1], c='blue', s=20, label='Inliers')
plt.scatter(outliers[:, 0], outliers[:, 1], c='red', s=20, label='Outliers', marker='x')
plt.legend()
plt.title("LOF Outlier Detection")
plt.show()
```
