```python
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
%matplotlib inline 
from sklearn.cluster import DBSCAN 
```

```python
X,y = make_circles(n_samples = 750,factor = 0.3 , noise = 0.1)
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
dbscan = DBSCAN(eps=0.10)
```

```python
predictions = dbscan.fit_predict(X)
predictions
```

```python
import numpy as np
index = np.where(predictions<0)
index
```

```python
plt.scatter(X[:,0],X[:,1],c=predictions) ##dbscab.labels__ can also be used besides predictions variable 
plt.scatter(X[index,0],X[index,1], c ='r') ##outlier denoted by red circles
```
