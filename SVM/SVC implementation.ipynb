## SVM Classifier implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,
                        n_clusters_per_class=1,n_redundant=0,random_state= 43)
```

```python
pd.DataFrame(X)
```

```python
X[:,1]
```

```python
y
```

```python
pd.DataFrame(X).iloc[:,0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue= y)
```

```python
from sklearn.svm import SVC

svc = SVC(kernel = 'linear') ##when data points are linearly separable use linear kernel
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.25, random_state= 10)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
svc.fit(X_train,y_train)
```

```python
##Predictions

y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

accuracy_score(y_test,y_pred) ##It will give good accuracy as we have good linearly separable data
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

dispcm =ConfusionMatrixDisplay(confusion_matrix = cm,display_labels = svc.classes_)
dispcm.plot()
```

### If we increase n_clusters

```python
from sklearn.datasets import make_classification
X1,y1=make_classification(n_samples=1000,n_features=2,n_classes=2,
                        n_clusters_per_class=2,n_redundant=0,random_state= 42)
```

```python
sns.scatterplot(x=pd.DataFrame(X).iloc[:,0],y=pd.DataFrame(X).iloc[:,1],hue= y)
```

```python
from sklearn.model_selection import train_test_split

X1_train,X1_test,y1_train,y1_test = train_test_split(X1,y1,test_size= 0.25, random_state= 10)
```

```python
svc1= SVC(kernel = 'linear')
svc1.fit(X1_train,y1_train)
```

```python
##Predictions

y1_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

accuracy_score(y1_test,y1_pred) ##It will give less accuracy as we dont have good linearly separable data now
```

```python
print(classification_report(y1_test,y1_pred)) ## Accuracy and all other metrics decreased
```

```python
print(confusion_matrix(y1_test,y1_pred))
```

```python
svc1.coef_ ##weights we calculated from fitting the line as we have 2 features so two slope or weights or coefficients
```

### In this case we try with Radial Kernel to deal with non linear data

```python
rbf = SVC(kernel= 'rbf')
```

```python
rbf.fit(X1_train,y1_train)
```

```python
##Predictions

y2_pred = rbf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

accuracy_score(y1_test,y2_pred) ##It will improve the accuracy as we used Radial kernel to deal with non linearl data that couldnt be separated by a line
```

```python
print(classification_report(y1_test,y1_pred)) ## Accuracy and all other metrics improved from linear kernel
```

```python
cm =confusion_matrix(y1_test, y2_pred)
cm
```

```python
displaycm = ConfusionMatrixDisplay(confusion_matrix= cm , display_labels= rbf.classes_)
displaycm.plot()
```

### Use polynomial Kernel

```python
poly = SVC(kernel = 'poly')

poly.fit(X1_train,y1_train)
```

```python
y3_pred= poly.predict(X1_test)
```

```python
accuracy_score(y1_test,y3_pred) ##It is giving worse accuracy than Radial a bit better than Linear but still not suitable for highly non linear data
```

```python
print(classification_report(y1_test,y3_pred))
```

```python
print(confusion_matrix(y1_test,y3_pred))
```

### Sigmoid Kernel

```python
sigm = SVC(kernel= 'sigmoid')
sigm.fit(X1_train,y1_train)
```

```python
y4_pred = sigm.predict(X1_test)
```

```python
print(accuracy_score(y1_test,y4_pred)) ##Worst performer amongst all and its natural as sigmoid uses 0.5 threshold for probability which is not optimal for Non linear data
```

```python
sigm.intercept_
```

### Conclusion: Use Radial Kernel for Non linear data 

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid = GridSearchCV( SVC() , param_grid=  param_grid, refit = True, cv= 5, verbose = 3)
```

```python
grid.fit(X1_train,y1_train)
```

```python
grid.best_params_
```

```python
y5_pred= grid.predict(X1_test)
```

```python
accuracy_score(y1_test,y5_pred) ##Improved accuracy due to best parameters selected
```

```python
print(classification_report(y1_test,y5_pred))
```

```python
confusion_matrix(y1_test,y5_pred)
```
