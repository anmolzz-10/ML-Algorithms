```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
import seaborn as sns
```

```python
data  = sns.load_dataset('iris')
```

```python
data
```

```python
df = load_iris()
```

```python
df
```

```python
df = pd.DataFrame(df.data,columns = df.feature_names)
```

```python
df
```

```python
##Standardization
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()
```

```python
X_scaled =  scaler.fit_transform(df)
```

```python
pd.DataFrame(X_scaled)
```

```python
## Apply pca

from sklearn.decomposition import PCA

pca= PCA(n_components=2)
```

```python
pca_scaled = pca.fit_transform(X_scaled)
pca_scaled
```

```python
data['species'].value_counts()
```

```python
data['species'] = data['species'].map({'setosa':0,'versicolor': 1,'virginica': 2})
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c= data['species'] )
```

```python
##Agglomerative Clustering
##To construct a dendogram

import scipy.cluster.hierarchy as sc

##plot the dendogram

plt.figure(figsize=(20,7))
plt.title('Dendograms')

sc.dendrogram(sc.linkage(pca_scaled,method= 'ward'))
plt.xlabel('Sample Index')
plt.ylabel('Euclidean Distance')
```

```python
from sklearn.cluster import *
```

```python
cluster = AgglomerativeClustering(n_clusters= 2 , affinity='euclidean', linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c= cluster.labels_)
```

```python
##We can observe that we have 2 clusters contrary to our original data that had 3 clusters
##This was found by dendrograms when a horizontal line is drawn to the longest vertical line, we get 2 points thus 2 clusters.
```
